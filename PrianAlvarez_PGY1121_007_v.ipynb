```python
import pickle

terreno = [[{'Estado': 'Vacío', 'Descripción': '', 'Precio': 5000000, 'Tamaño': "100 x 50"} for _ in range(5)] for _ in range(4)]

descripciones = ['Lote 1', 'Lote 2', 'Lote 3', 'Lote 4', 'Lote 5']


registros_ventas = {}

def mostrar_terreno():
    print("Estado del terreno:")
    for fila in terreno:
        for lote in fila:
            if lote['Estado'] == 'Vacío':
                print('[ ]', end=' ')
            else:
                print('[X]', end=' ')
        print()

def seleccionar_lote(fila, columna, cliente):
    try:
        lote = terreno[fila][columna]
        if lote['Estado'] == 'Vacío':
            lote['Estado'] = 'Seleccionado'
            lote_seleccionado = descripciones[fila]
            registros_ventas[lote_seleccionado] = cliente
            print(f'Has seleccionado el {lote_seleccionado}. Registro de venta guardado.')
        else:
            print('El lote', descripciones[fila], 'ya ha sido seleccionado.')
    except IndexError:
        print('Error: Las coordenadas del lote son inválidas.')

def detalles_lote(fila, columna):
    try:
        lote = terreno[fila][columna]
        lote_seleccionado = descripciones[fila]
        cliente = registros_ventas.get(lote_seleccionado)
        if cliente:
            print(f'Detalles del {lote_seleccionado}:')
            print(f'Descripción: {lote["Descripción"]}')
            print(f'Precio: ${lote["Precio"]}')
            print(f'Tamaño: {lote["Tamaño"]}')
            print(f'Cliente: {cliente}')
        else:
            print(f'El {lote_seleccionado} no ha sido seleccionado por ningún cliente.')
    except IndexError:
        print('Error: Las coordenadas del lote son inválidas.')

def mostrar_clientes():
    print("Clientes:")
    for lote, cliente in registros_ventas.items():
        print(f'Lote: {lote}, Cliente: {cliente}')

def guardar_registros():
    try:
        with open('registros_ventas.pickle', 'wb') as file:
            pickle.dump(registros_ventas, file)
    except IOError:
        print('Error al guardar los registros de ventas.')

def cargar_registros():
    global registros_ventas
    try:
        with open('registros_ventas.pickle', 'rb') as file:
            registros_ventas = pickle.load(file)
    except FileNotFoundError:
        registros_ventas = {}
    except IOError:
        print('Error al cargar los registros de ventas.')

cargar_registros()

def mostrar_menu():
    while True:
        print("\n--- Menú de opciones ---")
        print("1. Ver disponibilidad de lotes")
        print("2. Seleccionar un lote")
        print("3. Ver detalles del lote seleccionado")
        print("4. Ver Clientes")
        print("5. Salir")

        opcion = input("Ingrese el número de opción: ")

        if opcion == "1":
            mostrar_terreno()
        elif opcion == "2":
            try:
                fila = int(input("Ingrese el número de fila del lote: "))
                columna = int(input("Ingrese el número de columna del lote: "))
                cliente = input("Ingrese el nombre del cliente: ")
                seleccionar_lote(fila, columna, cliente)
                guardar_registros()
            except ValueError:
                print('Error: Ingrese valores numéricos válidos.')
        elif opcion == "3":
            try:
                fila = int(input("Ingrese el número de fila del lote: "))
                columna = int(input("Ingrese el número de columna del lote: "))
                detalles_lote(fila, columna)
            except ValueError:
                print('Error: Ingrese valores numéricos válidos.')
        elif opcion == "4":
            mostrar_clientes()
        elif opcion == "5":
          print('Hasta pronto :)')
          break
        else:
            print("Opción inválida. Por favor, ingrese un número válido.")

mostrar_menu()
```
